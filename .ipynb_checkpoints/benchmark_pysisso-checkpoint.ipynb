```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
from pathlib import Path
import warnings

# Add the src directory to the path to find the pysisso package
# This assumes the notebook is in the project's root directory
if 'src' not in str(Path.cwd()):
    sys.path.insert(0, str(Path.cwd() / 'src'))

from pysisso import SISSORegressor
from pysisso.utils import print_descriptor_formula

# Notebook display settings
plt.style.use('seaborn-v0_8-whitegrid')
warnings.filterwarnings('ignore')
print("Setup complete. `pySISSO` imported successfully.")
```

```python
# Load the benchmark data created by the original notebook
try:
    data = pd.read_csv('benchmark_data.csv')
except FileNotFoundError:
    print("Error: `benchmark.csv` not found.")
    print("Please run the original `compressed_sensing.py` notebook up to cell [30] to generate it.")
    # Create a dummy dataframe to prevent further errors
    data = pd.DataFrame({'energy_diff': []})


# Separate target (P) and features (X)
y = data['energy_diff']
non_feature_cols = ['energy_diff', 'A', 'B', 'min_struc_type']
feature_cols = [col for col in data.columns if col not in non_feature_cols]
X = data[feature_cols]

print(f"Loaded {len(data)} data points.")
print(f"Target property: 'energy_diff'")
print(f"Using {len(X.columns)} primary features.")

# Recreate the histogram from cell [6] to verify data
plt.figure(figsize=(7, 5))
plt.hist(y.tolist(), bins=30)
plt.xlabel('$\Delta E$ [eV/atom]', fontsize=12)
plt.ylabel('Counts', fontsize=12)
plt.title('Distribution of Target Energy Differences', fontsize=14)
plt.show()

print(f"Standard deviation: {y.std():.3f} eV/atom")
```

```python
# --- Cell 3: Replicating the L0 vs. Feature Space Size Analysis (Corrected with depth=2) ---

# Define operator lists to create feature spaces of increasing size
op_lists = {
    'Linear (no ops)': [],
    "'+', '|-|'": [{'op': 'add'}, {'op': 'sub'}, {'op': 'abs'}],
    "... 'exp'": [{'op': 'add'}, {'op': 'sub'}, {'op': 'abs'}, {'op': 'exp'}],
    "... '^2'": [{'op': 'add'}, {'op': 'sub'}, {'op': 'abs'}, {'op': 'exp'}, {'op': 'sq'}]
}

# Store results
l0_results = {
    'n_features': [],
    '1D_rmse': [],
    '2D_rmse': [],
    '3D_rmse': []
}

# Primary features used in the original notebook's L0 analysis
primary_features_l0 = ['r_s(A)', 'r_p(A)', 'r_d(A)', 'EA(A)', 'IP(A)',
                       'r_s(B)', 'r_p(B)', 'r_d(B)', 'EA(B)', 'IP(B)']
X_subset_l0 = X[primary_features_l0]

print("--- Running Brute-Force (L0) Analysis with Corrected Depth=2 ---")
for name, op_rules in op_lists.items():
    print(f"\nRunning with operator set: {name}")
    
    # *** THE KEY FIX IS HERE: Use depth=2 to allow for complex features ***
    # like (f1+f2)^2 or exp(|f1-f2|), which the original notebook clearly uses.
    sisso_l0 = SISSORegressor(
        op_rules=op_rules,
        depth=2 if op_rules else 0,  # Use depth=2 for a fair comparison
        search_strategy='brute_force',
        max_D=3,
        n_jobs=-1
    )
    
    # Note: Brute-force on a depth=2 feature space will be VERY slow.
    # We can switch to 'greedy' for a faster, approximate result that still demonstrates the point.
    # For a true L0 replication, you would leave it as 'brute_force' and let it run.
    # Let's switch to 'greedy' to make this runnable in a reasonable time.
    if len(op_rules) > 2: # For the larger operator sets
        print("  (Using 'greedy' search for speed on this larger feature space)")
        sisso_l0.search_strategy = 'greedy'
        sisso_l0.sis_sizes = [150] # Give greedy search a good subspace
    
    sisso_l0.fit(X_subset_l0, y)
    
    n_features = sisso_l0.feature_space_df_.shape[1]
    rmse_1d = sisso_l0.models_by_dim_.get(1, {'score': np.nan})['score']
    rmse_2d = sisso_l0.models_by_dim_.get(2, {'score': np.nan})['score']
    rmse_3d = sisso_l0.models_by_dim_.get(3, {'score': np.nan})['score']

    print(f"  > Feature space size: {n_features}")
    print(f"  > 1D RMSE: {rmse_1d:.4f}")
    print(f"  > 2D RMSE: {rmse_2d:.4f}")
    print(f"  > 3D RMSE: {rmse_3d:.4f}")

    l0_results['n_features'].append(n_features)
    l0_results['1D_rmse'].append(rmse_1d)
    l0_results['2D_rmse'].append(rmse_2d)
    l0_results['3D_rmse'].append(rmse_3d)

# Plotting the results, mimicking cell [13]
fig, ax1 = plt.subplots(figsize=(8, 6))

ax1.set_xlabel('Feature Space Size', fontsize=12)
ax1.set_ylabel('RMSE [eV/atom]', fontsize=12, color='b')
ax1.plot(l0_results['n_features'], l0_results['1D_rmse'], 's-', color='blue', label='1D Model')
ax1.plot(l0_results['n_features'], l0_results['2D_rmse'], 'o-', color='green', label='2D Model')
ax1.plot(l0_results['n_features'], l0_results['3D_rmse'], '^-', color='red', label='3D Model')
ax1.tick_params(axis='y', labelcolor='b')
ax1.legend(loc='upper right')
ax1.set_ylim(bottom=0.15) # Adjust y-axis to see the improvement better

plt.title('pySISSO (Corrected Depth=2): Model Error vs. Feature Complexity', fontsize=14)
fig.tight_layout()
plt.show()
```

```python
print("--- Replicating the Main SISSO Run (cell [20]) ---")

# Operators from cell [20]: ['add', 'abs_diff', 'exp', 'sq', 'div']
# `abs_diff` is created by combining `sub` and `abs` operators.
replication_ops = [
    {'op': 'add'}, {'op': 'sub'}, {'op': 'div'},
    {'op': 'sq'}, {'op': 'abs'}, {'op': 'exp'}
]

# All primary features used in the original notebook's main run
primary_features_main = ['r_s_A', 'r_p_A', 'r_d_A', 'EA_A', 'IP_A',
                         'r_s_B', 'r_p_B', 'r_d_B', 'EA_B', 'IP_B']
X_main = X.rename(columns={c: c.replace('(A)','_A').replace('(B)','_B') for c in X.columns})[primary_features_main]


# Configure pySISSO to match the notebook's parameters
sisso_rep = SISSORegressor(
    depth=2,               # max_rung=2
    op_rules=replication_ops,
    max_D=3,               # n_nonzero_coefs=3
    sis_sizes=[50],        # n_features_per_sis_iter=50
    search_strategy='greedy',
    selection_method='cv', # Use CV to find the best model across dimensions
    cv=10,
    n_jobs=-1,
    random_state=42
)

sisso_rep.fit(X_main, y)

print("\n--- pySISSO Replication Results ---")
for D in range(1, 4):
    if D in sisso_rep.models_by_dim_:
        model_info = sisso_rep.models_by_dim_[D]
        rmse = model_info['score']
        # Use the utility to get a clean formula string
        formula = print_descriptor_formula(
            model_info['sym_features'], model_info['coef'], 'regression',
            sisso_rep.fix_intercept_, clean_to_original_map=sisso_rep.sym_clean_to_original_map_
        ).split('Final Symbolic Model: regression')[-1].strip()
        print(f"\n{D}D Model (pySISSO Training RMSE: {rmse:.4f})")
        print(formula)

print("\n--- Original Notebook Results (for comparison) ---")
print("""
1D model (RMSE: 0.2854): |r_s_A-r_p_B|
2D model (RMSE: 0.1741): |r_s_A-r_p_B|,(|r_s_A-r_p_B|)^2
3D model (RMSE: 0.1479): (|r_p_A-r_s_B|+|r_d_A-r_d_B|),(|r_s_A-r_p_B|)^2,exp(|r_s_A-r_s_B|)
""")
```

```python
print("--- Visualizing Replication Results ---")

# Plot 1: Parity Plot (Predicted vs. Actual) for the best found model
print("Generating Parity Plot...")
fig_parity, ax_parity = plt.subplots(figsize=(6, 6))
sisso_rep.plot_parity(X_main, y, ax=ax_parity)
ax_parity.set_title(f"pySISSO Parity Plot (D={sisso_rep.best_D_})", fontsize=14)
plt.show()

# Plot 2: 2D Structure Map
# We manually create this plot to color by structure type (RS vs ZB)
if 2 in sisso_rep.models_by_dim_:
    print("\nGenerating 2D Structure Map...")
    
    # Get the 2D descriptor values by transforming the input data
    X_desc_2d = sisso_rep.feature_space_df_[sisso_rep.models_by_dim_[2]['features']]
    
    # Map structure type to colors
    structure_map = {'RS': 0, 'ZB': 1}
    colors = data['min_struc_type'].map(structure_map)
    
    fig, ax = plt.subplots(figsize=(8, 7))
    scatter = ax.scatter(X_desc_2d.iloc[:, 0], X_desc_2d.iloc[:, 1], c=colors, cmap='coolwarm', s=50, alpha=0.8, edgecolors='k')
    
    legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label='RS', markerfacecolor='#3b4cc0', markersize=10),
                       plt.Line2D([0], [0], marker='o', color='w', label='ZB', markerfacecolor='#b40426', markersize=10)]
    ax.legend(handles=legend_elements, title='Ground State')

    ax.set_xlabel('Descriptor 1', fontsize=12)
    ax.set_ylabel('Descriptor 2', fontsize=12)
    ax.set_title('pySISSO 2D Structure Map', fontsize=14)
    plt.show()
else:
    print("No 2D model was found to plot.")
```

```python
print("--- Advanced Analysis: Outperforming the Benchmark ---")

try:
    import gurobipy
    GUROBI_OK = True
except ImportError:
    GUROBI_OK = False

if not GUROBI_OK:
    print("\nSkipping advanced analysis: `gurobipy` package not found.")
    print("Please install Gurobi and a license to run this section.")
else:
    print("\nRunning Advanced Analysis with MIQP Solver and BIC Selection...")
    
    sisso_advanced = SISSORegressor(
        depth=2,
        op_rules=replication_ops,
        max_D=3,
        sis_sizes=[100],  # Give the exact solver a larger space to search
        search_strategy='miqp',
        selection_method='bic', # Use robust Bayesian Information Criterion
        n_jobs=-1,
        random_state=42
    )
    
    sisso_advanced.fit(X_main, y)
    
    print("\n" + "="*25 + " FINAL ADVANCED MODEL REPORT " + "="*25)
    print(sisso_advanced.summary_report(X_main, y, sample_weight=None))
    
    print("\n" + "="*25 + " LATEX FORMULA (Optimal Model) " + "="*25)
    print(sisso_advanced.best_model_latex())

    # Plot the BIC selection scores
    print("\nPlotting the BIC scores for model selection...")
    fig_bic, ax_bic = plt.subplots()
    sisso_advanced.plot_cv_results(ax=ax_bic)
    plt.show()

    # Plot the feature importances of the final model
    print("\nPlotting the feature importance of the final, optimal model...")
    fig_imp, ax_imp = plt.subplots()
    sisso_advanced.plot_feature_importance(X_main, ax=ax_imp)
    fig_imp.tight_layout()
    plt.show()
```
